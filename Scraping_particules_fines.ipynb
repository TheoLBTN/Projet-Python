```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://fr.wikipedia.org/wiki/Norme_europ%C3%A9enne_d%27%C3%A9missions"

# Envoi de la requête
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Recherche du tableau pour "Diesel"
diesel = soup.find("span", {"id": "V.C3.A9hicules_.C3.A0_moteur_Diesel"}).find_next("table")

# Recherche du tableau pour "Essence"
essence = soup.find("span", {"id": "V.C3.A9hicules_.C3.A0_moteur_essence_et_GPL.2FGNL"}).find_next("table")

euro = soup.find('table', {'class' : 'wikitable'})

# Conversion des tableaux en DataFrames
df_diesel = pd.read_html(str(diesel))[0]
df_essence = pd.read_html(str(essence))[0]
```

```python
df_diesel = df_diesel.drop([0, 1, 9], axis = 0)
df_diesel = df_diesel.reset_index()
df_diesel = df_diesel.drop(['index'], axis = 1)
```

```python
df_essence = df_essence.drop([7, 8], axis = 0)
```

```python
df_euro=pd.read_html(str(euro))[0]
df_euro=df_euro.drop(["Homologation des nouveaux types"], axis=1)
```

```python
df_diesel.to_csv('emissions_diesel.csv', index = False)
df_essence.to_csv('emissions_essence.csv', index = False)
df_euro.to_csv('norme_euro.csv', index=False)
```
